```python
#added truncation
import tqdm

# Dependency parsing
sentences1 = list(df_to_save['sentence1'])
sentences2 = list(df_to_save['sentence2'])

# Maximum sequence length
max_length = 100

# Truncate a sentence string to the specified length
def truncate_sentence(sentence, max_length):
    words = sentence.split()
    truncated_words = words[:max_length]
    return ' '.join(truncated_words)

# Wrap the truncated sentences with stanza.Document objects
in_docs1 = [stanza.Document([], text=truncate_sentence(d, max_length)) for d in sentences1]
in_docs2 = [stanza.Document([], text=truncate_sentence(d, max_length)) for d in sentences2]

# Set batch size
batch_size = 64

# Process the sentences in batches using the Stanza pipeline
def process_batches(in_docs):
    processed_docs = []
    for i in tqdm.tqdm(range(0, len(in_docs), batch_size), desc="Processing batches"):
        batch = in_docs[i:i + batch_size]
        processed_batch = nlp(batch)
        processed_docs.extend(processed_batch)
    return processed_docs

docs1 = process_batches(in_docs1)
docs2 = process_batches(in_docs2)
```

```python
text1, text2 = [],[]
postags1, postags2 = [],[]
heads1, heads2 = [],[]
deprel1, deprel2 =[],[]
const1,const2=[],[]

for sentences in docs1:
  text1.append([x['text'] for x in sentences.sentences[0].to_dict()])
  postags1.append([x['xpos'] for x in sentences.sentences[0].to_dict()])
  heads1.append([x['head'] for x in sentences.sentences[0].to_dict()])
  deprel1.append([x['deprel'] for x in sentences.sentences[0].to_dict()])
  const1.append(str(sentences.sentences[0].constituency))

for sentences in docs2:
  text2.append([x['text'] for x in sentences.sentences[0].to_dict()])
  postags2.append([x['xpos'] for x in sentences.sentences[0].to_dict()])
  heads2.append([x['head'] for x in sentences.sentences[0].to_dict()])
  deprel2.append([x['deprel'] for x in sentences.sentences[0].to_dict()])
  const2.append(str(sentences.sentences[0].constituency))

# assert len(text1)==len(deprel1) == len(heads1) == len(postags1) == len(df_to_save)
assert len(text2)==len(deprel2) == len(heads2) ==len(postags2) == len(df_to_save)
```

```python
df_to_save['text_sentence1'] = text1
df_to_save['text_sentence2']=text2
df_to_save['pos_sentence1']=postags1
df_to_save['pos_sentence2']=postags2
df_to_save['heads_sentence1']= heads1
df_to_save['heads_sentence2']= heads2
df_to_save['deprel_sentence1']=deprel1
df_to_save['deprel_sentence2']=deprel2
df_to_save['sentence1_parse']=const1
df_to_save['sentence2_parse']=const2
```
